```python
import matplotlib.pyplot as plt

# Exemples de données de perte par époque
epochs = list(range(1, 101))  # 100 époques
train_loss = [1 / (0.1 * epoch + 1) + 0.05 * (epoch % 5) for epoch in epochs]
val_loss = [1 / (0.1 * epoch + 1.2) + 0.05 * ((epoch + 2) % 5) for epoch in epochs]

# Création de la figure
plt.figure(figsize=(10, 6))
plt.plot(epochs, train_loss, label='Perte Entraînement', color='blue')
plt.plot(epochs, val_loss, label='Perte Validation', color='orange')
plt.title("Figure 1 : Évolution de la fonction de perte (train vs validation)")
plt.xlabel("Époques")
plt.ylabel("Perte")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("image_01.png")  # Enregistre la figure
plt.show()
```

```python
import matplotlib.pyplot as plt

# Simuler des précisions par époque (ou les récupérer via un modèle réel)
epochs = list(range(1, 101))  # 100 époques
train_accuracy = [0.6 + 0.004 * epoch - 0.005 * (epoch % 7) for epoch in epochs]
val_accuracy = [0.58 + 0.004 * epoch - 0.004 * ((epoch + 3) % 6) for epoch in epochs]

# Création de la figure
plt.figure(figsize=(10, 6))
plt.plot(epochs, train_accuracy, label='Précision Entraînement', color='green')
plt.plot(epochs, val_accuracy, label='Précision Validation', color='red')
plt.title("Figure 2 : Évolution de la précision au fil des époques (train vs validation)")
plt.xlabel("Époques")
plt.ylabel("Précision")
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("image_02.png")  # Sauvegarde la figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np

# --- Exemple fictif (à remplacer par tes vrais labels) ---
y_true = [0, 1, 2, 2, 1, 0, 1, 2, 0, 1] * 10
y_pred = [0, 2, 2, 2, 1, 0, 1, 0, 0, 1] * 10

# Calcul de la matrice de confusion
cm = confusion_matrix(y_true, y_pred)
labels = [f"C{i}" for i in range(len(cm))]  # Exemple : C0, C1, C2

# Affichage plus petit et avec une palette différente
plt.figure(figsize=(6, 5))  # Taille ajustée ici
sns.heatmap(cm, annot=True, fmt='d', cmap="Purples", 
            xticklabels=labels, yticklabels=labels, linewidths=0.5, linecolor='gray')

plt.title("Figure 3 : Matrice de confusion", fontsize=11)
plt.xlabel("Prédit", fontsize=10)
plt.ylabel("Réel", fontsize=10)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
